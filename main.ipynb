## O IDH brasileiro melhorou? Foi um crescimento uniforme nos três componentes?

```python
amnt_null_idh = df_idh_mun[df_idh_mun['IDHM'].isnull()].count()[0]
amnt_null_educ = df_idh_mun[df_idh_mun['IDHM_E'].isnull()].count()[0]
amnt_null_longev = df_idh_mun[df_idh_mun['IDHM_L'].isnull()].count()[0]
amnt_null_renda = df_idh_mun[df_idh_mun['IDHM_R'].isnull()].count()[0]
amnt_null_idh + amnt_null_educ + amnt_null_longev + amnt_null_renda
```

*Não há nenhum valor nulo no DataFrame nas colunas IDHM, IDHM_E, IDHM_R, IDHM_L*

```python
df_idh_mun[df_idh_mun['IDHM']==0].count()[0]
null_idh = df_idh_mun.count()[0] - df_idh_mun[df_idh_mun['IDHM'].notnull()].count()[0]
null_educ = df_idh_mun.count()[0] - df_idh_mun[df_idh_mun['IDHM_E'].notnull()].count()[0]
null_renda = df_idh_mun.count()[0] - df_idh_mun[df_idh_mun['IDHM_R'].notnull()].count()[0]
null_longev = df_idh_mun.count()[0] - df_idh_mun[df_idh_mun['IDHM_L'].notnull()].count()[0]

null_educ + null_renda + null_educ + null_idh
```

```python
gp_idh_mean_year = df_idh_mun.groupby('ANO')[['IDHM', 'IDHM_R', 'IDHM_E', 'IDHM_L']].mean()
gp_idh_mean_year.columns = ['IDH', 'renda', 'educ', 'longev']
gp_idh_mean_year.head()
```

```python
gp_idh_mean_year[['IDH']].plot()
gp_idh_mean_year[['renda', 'educ', 'longev']].plot()
```

Podemos observar que o IDH brasileiro melhorou durante os três últimos censos se aproximando muito de uma reta e o indicador que mais aumentou nesse período foi o de educação.

```python
df_idh_mean_year = gp_idh_mean_year.reset_index()

idh_1991 = df_idh_mean_year[df_idh_mean_year['ANO']==1991]
idh_2010 = df_idh_mean_year[df_idh_mean_year['ANO']==2010]
```

```python
data_idh_increased_abs = {
    'aumento abs IDH':  [idh_2010['IDH'].item() - idh_1991['IDH'].item()],
    'aumento abs Renda':  [idh_2010['renda'].item() - idh_1991['renda'].item()],
    'aumento abs Educacao':  [idh_2010['educ'].item() - idh_1991['educ'].item()],
    'aumento abs Longevidade':  [idh_2010['longev'].item() - idh_1991['longev'].item()],
}
pd.DataFrame(data_idh_increased_abs)
```

```python
data_idh_increased_abs = {
    'aumento IDH (%)': [(idh_2010['IDH'].item()/idh_1991['IDH'].item()-1)*100],
    'aumento Renda (%)': [(idh_2010['renda'].item()/idh_1991['renda'].item()-1)*100],
    'aumento Educacao (%)': [(idh_2010['educ'].item()/idh_1991['educ'].item()-1)*100],
    'aumento Longevidade (%)': [(idh_2010['longev'].item()/idh_1991['longev'].item()-1)*100],
}
pd.DataFrame(data_idh_increased_abs)
```

Novamente podemos observar que o indicador que teve mais aumento absoluto e percentual foi o de educação. Os outros 2 tiveram um aumento mediano e bem próximos.
Dessa forma, o que mais influenciou o crescimento do IDH nesse período foi a melhoria na educação.
Mesmo assim, esse componente ainda é o menor dentre os três.

## Há uma relação entre Renda e Educação?

A coluna UFN (nome de cada estado) não está presente no dataframe como dizia então busquei um dataframe que contém a relação de código dos estados, nome e sigla. Então podemos fazer um merge dos dataframes.

```python
PATH_CODIGO_ESTADOS = 'Datasets/Codigos-UF/codigos_uf.csv'
df_codigo_estados = pd.read_csv(PATH_CODIGO_ESTADOS, sep=r'\s*,\s*', header=0, engine='python')
df_codigo_estados.head()
```

```python
df_idh_mun['UFN'] = df_idh_mun['UF'].map(lambda codigo: df_codigo_estados[df_codigo_estados['Codigo']==codigo]['UFN'].item())
df_idh_mun['UF'] = df_idh_mun['UF'].map(lambda codigo: df_codigo_estados[df_codigo_estados['Codigo']==codigo]['UF'].item())
df_idh_mun[['UF', 'UFN']].head()
```

```python
df_idh_est = df_idh_mun[df_idh_mun['ANO']==2010].groupby('UF')[['IDHM', 'IDHM_R', 'IDHM_E', 'IDHM_L']].mean()
df_idh_est.head()
```

```python
df_idh_correl = df_idh_est[['IDHM_E', 'IDHM_R']].sort_values('IDHM_R')
df_idh_correl.plot.bar(figsize=(15, 10))
```

```python
df_idh_correl.head()
```

```python
import seaborn as sns
```

```python
sns.scatterplot(df_idh_correl['IDHM_E'], df_idh_correl['IDHM_R'])
```

Podemos observar que esses pontos seguem, mesmo que de forma não tão forte e precisa, uma reta crescente. Levando a entender que as duas medidas se comportam de maneira diretamente proporcional.

Para confirmar essa suposição vamos utilizar a função correlation pois estamos lidando com dois dados quantitativos.

# Covid
Dados do covid no Brasil. Datasets:
- Dados de casos acumulados e óbtidos por município e por dia. Disponível em: https://covid.saude.gov.br/.
- Dados dos pacientes de Covid19. Disponível em: https://brasil.io/dataset/covid19/caso_full/.

Os dados são os mais atualizados até 19 de maio.

```python
PATH_COVID_CASOS = 'Datasets/covid19/HIST_PAINEL_COVIDBR_19mai2020.xlsx'
df_casos = pd.read_excel(pd.ExcelFile(PATH_COVID_CASOS))
df_casos.head()
```

```python
df_casos['coduf'].notnull().mean()
```

Temos algumas linhas sem município e codmun, porém todos as linhas tem a informação do estado pelo coduf.

```python
df_cod_estados = df_casos[df_casos['estado'].notnull()][['coduf', 'estado']].groupby('coduf').first()
df_cod_estados = df_cod_estados.reset_index()
df_cod_estados
```

```python
df_casos.groupby('coduf')['estado'].agg(['first', 'count'])
```

Por algum motivo há um código 76 que não existe, então vamos removê-lo.

```python
df_casos = df_casos[df_casos['coduf']!=76]
```

Agora vamos adicionar o nome dos estados a todos os que estão com o nome nulo.

```python
estado_from_cod = lambda cod : df_cod_estados[df_cod_estados.where(df_cod_estados['coduf']==cod)['estado'].notnull()]['estado'].item()
estado_from_cod(53) #teste
```

```python
df_casos = df_casos.reset_index()
del df_casos['index']
df_casos.head()
df_casos.at[1,"estado"]
for row in range(0, len(df_casos)) :
    df_casos.at[row, "estado"] = estado_from_cod(df_casos.at[row, "coduf"])
```

Todos os municípios estão com os dados atualizados até o mesmo dia?

```python
df_casos_mun = df_casos[df_casos['codmun'].notnull()]

from functools import reduce
def ultimaAtualizacao(series) :
    return reduce(lambda x,y: x if x>y else y, series)
last_update_mun = df_casos_mun.groupby('codmun').agg({'data': [ultimaAtualizacao]})
last_update_mun = last_update_mun.reset_index()
last_update_mun
```

```python
last_update_mun['ultimaAtualizacao'] = last_update_mun[('data', 'ultimaAtualizacao')]
del last_update_mun[('data', 'ultimaAtualizacao')]
last_update_mun.groupby('ultimaAtualizacao').count()
```

Praticamente todas as cidades foram atualizadas pela última vez no dia 19/05, então não precisaremos desconsiderar nenhum dia.

Vamos analisar agora a frequência de rows por dia por município.

```python
gb_data_mun = df_casos_mun[['casosAcumulado', 'data', 'codmun']].copy()
gb_data_mun['group'] = gb_data_mun['codmun'].astype(str) + gb_data_mun['data']
df_data_mun = gb_data_mun.groupby('group').agg({'data': ['count']}).reset_index()

df_data_mun['count'] = df_data_mun[('data', 'count')]
del df_data_mun[('data', 'count')]
df_data_mun
```

```python
df_data_mun['all'] = True
df_data_mun.groupby('all').max()
```

Todos os municípios tem no máximo 1 row por dia.
